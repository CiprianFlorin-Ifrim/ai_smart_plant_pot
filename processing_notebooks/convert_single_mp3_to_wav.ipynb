```python
import os

# Add the path to ffmpeg and ffprobe before importing pydub
ffmpeg_path = '/opt/homebrew/bin'  # Common path for Homebrew installations on Apple Silicon Macs
if os.path.exists(ffmpeg_path):
    os.environ["PATH"] += os.pathsep + ffmpeg_path

from pydub import AudioSegment
```

```python
def convert_mp3_to_wav(mp3_file_path, wav_file_path):
    # Load the MP3 audio file
    audio = AudioSegment.from_mp3(mp3_file_path)

    # Export the audio file as WAV
    audio.export(wav_file_path, format="wav")
    print(f"Converted {mp3_file_path} to {wav_file_path} successfully.")

def main():
    # Get the path to the MP3 file
    input_mp3 = input("Enter the path to the MP3 file: ")
    
    # Get the path for the output WAV file
    output_wav = input("Enter the output WAV file path: ")

    # Convert the MP3 to WAV
    convert_mp3_to_wav(input_mp3, output_wav)

if __name__ == "__main__":
    main()
```
